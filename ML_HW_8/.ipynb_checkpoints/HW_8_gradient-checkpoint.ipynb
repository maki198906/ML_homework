```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.simplefilter('ignore')
```

```python
data = load_iris()
```

```python
target_names = data['target_names'][1:]
```

```python
target_names
```

```python
X = pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
X['target'] = data['target']
```

```python
X = X[X['target'] !=0]
```

```python
# versicolor = 1, virginica = 0
```

```python
f = {1:1,2:0}
```

```python
X['target'] = X['target'].map(f)
```

```python
y = X['target']
```

```python
X.drop(columns='target', axis=1,inplace=True)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y)
```

Логистическая функция (иногда также называемая сигмоидом или логит-функцией)

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
from IPython.display import Image
Image(filename='cost_function_logreg.png')
```

```python
def grad_des(X, y, n_epochs=2000, learning_rate = 0.05):
    w = np.ones(X.shape[1]) # начальные коэфф
    m = X.shape[0] # кол-во элементов
    mse_cost = []
    for _ in range(n_epochs):
        z = np.dot(w,X.T) # сколярное произведение w-транспонированное на X
        h_x = sigmoid(z)
        # w = w - learning_rate * dw 
        dw = np.dot(X.T, (h_x-y)) * 1/X.shape[1] 
        w = w - learning_rate * dw
        J = (1/m) * np.sum((-y * np.log(h_x)) - ((1-y) * np.log(1-h_x))) # Loss-function
        mse_cost.append(J)
    return w, mse_cost
```

```python
w, mse_cost = grad_des(X_train,y_train)
```

посмотрим на график распределения Loss-function

```python
fig , ax = plt.subplots(1,1,figsize=(14,6))
ax.plot(mse_cost)
ax.set_xscale('log')
ax.set_xlim(-1000,1000)
ax.set_ylim(0,18)
plt.show()
```

сильные колебания вначале -> плавное затухание 

Посмотрим на confusion_matrix

```python
def prediction(x,w):
    z = np.dot(w,x.T)
    h_x = sigmoid(z)
    return np.round(h_x)
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
confusion_matrix(y_val,prediction(X_val,w))
```

не угадали лишь один раз, хороший результат

## nesterov momentum

```python
def nesterov_momentum (X, y, n_epochs=20000, learning_rate = 0.01, gamma = .95):
    w = np.ones(X.shape[1]) 
    m = X.shape[0]
    mse_cost = []
    grad_initial = np.zeros(X.shape[1])
    for _ in range(n_epochs):
        w_ = w - grad_initial
        z = np.dot(w_,X.T) 
        h_x = sigmoid(z)
        dw = np.dot(X.T, (h_x-y)) * 1/X.shape[1]
        grad = gamma*grad_initial + learning_rate*dw
        w = w - grad
        grad_initial = grad
        J = (1/m) * np.sum((-y * np.log(h_x)) - ((1-y) * np.log(1-h_x))) # Loss-function
        mse_cost.append(J)
    return w, mse_cost
```

```python
w, mse_cost = nesterov_momentum(X_train,y_train)
```

посмотрим на график распределения Loss-function

```python
fig , ax = plt.subplots(1,1,figsize=(14,6))
ax.plot(mse_cost)
ax.set_xscale('log')
ax.set_xlim(0,1000)
ax.set_ylim(0,6)
plt.show()
```

Колебания стали меньше

```python
confusion_matrix(y_val,prediction(X_val,w))
```

## RMS_Prop

```python
def rms_prop (X, y, n_epochs=25000, eta = 0.01, gamma = 0.99, eps = 1e-8):
    w = np.ones(X.shape[1]) 
    m = X.shape[0]
    grad_initial = np.zeros(X.shape[1])
    mse_cost = []
    for _ in range(n_epochs):
        z = np.dot(w,X.T)
        h_x = sigmoid(z)
        dw = np.dot(X.T, (h_x-y)) * 1/X.shape[1]
        grad_square = dw**2
        grad_square_avg = gamma * grad_initial + (1-gamma) * grad_square
        w = w - eta/np.sqrt(grad_square_avg + eps) * dw
        grad_initial = grad_square_avg
        J = (1/m) * np.sum((-y * np.log(h_x)) - ((1-y) * np.log(1-h_x))) # Loss-function
        mse_cost.append(J)
    return w, mse_cost
```

```python
w, mse_cost = rms_prop(X_train,y_train)
```

посмотрим на график распределения Loss-function

```python
fig , ax = plt.subplots(1,1,figsize=(14,6))
ax.plot(mse_cost)
ax.set_xscale('log')
ax.set_xlim(0,1000)
ax.set_ylim(0,8)
plt.show()
```

колебаний нет, плавные переходы

```python
confusion_matrix(y_val,prediction(X_val,w))
```

# Классификация (Теория)

## Cross-Entropy Loss

## Вероятностный взгляд. Бинарная классификация
Рассмотрим задачу классификации. Предположим, что ответ нашей модели $h_\theta(x_i)$ получен на основе логистической регрессии ($\sigma(Wx_i + b)$). Ее значения лежат в диапазоне от 0 до 1, что может быть интерпретировано, как вероятность, что $x_i$ принадлежит positive классу. Если вероятность $< 0.5$ - мы делаем предсказание о принадлежности отрицательному классу, а если $>= 0.5$ - положительному. Следовательно:

$$p(y_i = 1 \vert x_i)  = h_\theta(x_i)$$
$$p(y_i = 0 \vert x_i) = 1 - h_\theta(x_i)$$

Мы можем соединить это в одно уравнение:

$$p(y_i | x_i) = [h_\theta(x_i)]^{(y_i)} [1 - h_\theta(x_i)]^{(1 - y_i)}$$

Опять-таки из предположения, что наши данные независимы и одинаково распределены перейдем к правдоподобию:

$$L(x, y) = \prod_{i = 1}^{N}[h_\theta(x_i)]^{(y_i)} [1 - h_\theta(x_i)]^{(1 - y_i)}$$

Точно так же, как в случае MSE, возмьмем логарифм и инвертируем знак. Получим наш loss:

$$J = -\sum_{i=1}^{N} y_i\log (h_\theta(x_i)) + (1 - y_i)\log(1 - h_\theta(x_i))$$

## Многоклассовая классификация

Основная идея - у нас ничего не меняется. Просто теперь вместо двух классов надо учитывать вероятности нескольких классов и теперь наш лосс примет такой вид:

$$-\sum_{i=1}^{N} \sum_{j=1}^{K} y_{ij} \log(h_{\theta}(x_{i}){_j})$$

## Специальные виды cross_entropy
### sigmoid cross-entropy (Binary Cross-Entropy Loss)
Сначала применяем sigmoid, потом считаем cross_entropy. Применяется, когда объект может принадлежать нескольким классам.

$${\displaystyle \sigma (x)={\frac {1}{1+e^{-x}}}.}$$

### softmax cross-entropy (Categorical Cross-Entropy loss)
Сначала делаем softmax, потом считаем cross_entropy. Применяется, когда объект может принадлежать только одному классу. Т.к. softmax выравнивает вероятности (суммируются к одному для всех классов)

$${\displaystyle \sigma (z)_{i}={\frac {e^{z_{i}}}{\displaystyle \sum _{k\mathop {=} 1}^{K}e^{z_{k}}}}}$$
