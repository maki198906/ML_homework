## Попробуем стекинг

```python
def get_meta_train_test(models, X_train, X_val, y_train):
    
    X_train_meta = np.empty((y_train.shape[0], len(models)))
    X_meta_val = np.empty((X_val.shape[0], len(models)))
    stack_cv = KFold(n_splits=10, random_state=27)
    
    for train_id, test_id in stack_cv.split(X_train, y_train):
        X_train_cv, X_test_cv = X_train.iloc[train_id,:], X_train.iloc[test_id,:]
        y_train_cv, y_test_cv = y_train.iloc[train_id], y_train.iloc[test_id]
        
        for i, model in enumerate(models):
            model.fit(X_train_cv, y_train_cv)            
            X_train_meta[test_id, i] = model.predict(X_test_cv)
            
    for j, model in enumerate(models):
        model.fit(X_train,y_train)
        y_val_pred = model.predict(X_val)
        X_meta_val[:, j] = y_val_pred       
            
    return X_train_meta, X_meta_val
```

```python
meta_train_X, meta_val_X = get_meta_train_test(models,X_train,X_val,y_train)
```

```python
X_meta_train = pd.DataFrame(meta_train_X,columns=['RandomForestRegressor', 'Ridge','SVR','Lasso'])
X_meta_val = pd.DataFrame(meta_val_X,columns=['RandomForestRegressor', 'Ridge','SVR','Lasso'])
```

```python
y_pred_clf_liner_meta = clf_liner_meta.predict(X_meta_val)
```

```python
mean_squared_log_error(y_val,y_pred_clf_liner_meta)
```
